```python
cat_ins_sum = cleanData.groupby(['Category'])['Installs'].sum()
cat_ins_sum.plot(kind='bar',ylabel="Installs",color="green",figsize=(20,5))
```

```python
cat_ins_sum = cleanData.groupby(['Category'])['Rating'].mean()
cat_ins_sum.plot(kind='bar',ylabel="Rating",color="purple",figsize=(20,5))
```

```python
cat_ins_sum = cleanData.groupby(['Category'])['Reviews'].mean()
cat_ins_sum.plot(kind='bar',ylabel="Reviews",color="orange",figsize=(20,5))
```

```python
cat_ins_sum = cleanData.groupby(['Category'])['Price in Dollars'].mean()
cat_ins_sum.plot(kind='bar',ylabel="Price in Dollars",color="yellow",figsize=(20,5))
```

```python
cat_ins_sum = cleanData.groupby(['Genres'])['Installs'].sum()
cat_ins_sum.plot(kind='bar',ylabel="Installs",color="brown",figsize=(20,5))
```
