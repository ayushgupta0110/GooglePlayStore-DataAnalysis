```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import plotly.express as px
```

```python
def fetchReviews(appId):
  return reviews_all(
      appId,
      lang='en', # defaults to 'en'
      country='us', # defaults to 'us'
      sort=Sort.NEWEST, # defaults to Sort.NEWEST
  )
```

```python
appIds=['com.divinegames.weddingfashionmakeover','com.fashiondollmakeupgames.fashiongirlgames2020','com.fashiondollmakeupgames.fashiongirlgames2020','com.fashiondollmakeupgames.fashiongirlgames2020','com.yamtargames.snakesim','com.yamtargames.aliensim','com.kirito.app.wallpaper.anime','com.wallio.wallpaper','com.animewallpapers','id.akusantri.artdrawingideas','com.gotools.artistgrid','com.app.offerit','com.embibe.student','org.cambridge.cdo.plus','com.englishhindidictionary.offlinedictionary','com.ma.cc.indian','com.endless.indianrecipes','com.irapps.shareme.filetransfer.share.data','com.smartswitch.phoneclone.backuprestore.oldtonewphone.datatransfer.mobiledata.android.phoneclone','dating.find.hookup.apps.sugar.single.free','com.over.seniors','com.pdfconverter.imagetopdf','imagetopdf.pdfconverter.jpgtopdf.img2pdf.converter','in.healthandglow','com.adriaandejongh.hiddenfolks'];
```

```python
len(appIds)
```

```python
appsReviews=[];
for index,appId in enumerate(appIds):
  appsReviews.append(fetchReviews(appIds[index]));
  appsReviews[index]=appsReviews[index][:5];
  for i in range(len(appsReviews[index])):
    appsReviews[index][i]['appId']=appIds[index];
appsReviews
```

```python
appsReviewsJson=[];
for index1,app in enumerate(appsReviews):
  for index2,review in enumerate(appsReviews[index1]):
    appsReviewsJson.append(appsReviews[index1][index2]);
appsReviewsJson;
```

```python
df=pd.json_normalize(appsReviewsJson)
```

```python
df
```

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
print(sentiment_analysis("I love this!"))
```

```python
df.dtypes
```

```python
df['content']=df['content'].astype('string')
```

```python
df.dtypes
```

```python
df.shape
```

```python
df['result']=df['content'][:df.shape[0]].apply(lambda x: sentiment_analysis(x))
```

```python
df['sentiment']=df['result'][:df.shape[0]].apply(lambda x: x[0]['label'])
```

```python
df['sentiment_score']=df['result'][:df.shape[0]].apply(lambda x: x[0]['score'])
```
