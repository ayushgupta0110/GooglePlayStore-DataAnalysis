```python
from google_play_scraper import app;
import pandas as pd;
```

```python
appIds=['com.instagram.android','com.ludo.king','com.phonepe.app','com.accurate.live.weather.forecast.pro','com.phonepe.app','in.pricehistory.app','com.webgames.ghosts','com.sbi.SBIFreedomPlus','com.microsoft.office.officehubrow','com.netflix.mediaclient','com.jio.media.stb.ondemand','com.amazon.kindle','com.truecaller','com.google.earth','com.irctc.air','com.booking','com.snapbreak.strangercases','com.adobe.scan.android','com.playwithgames.CoastGuardParking','in.amazon.mShop.android.shopping','com.nisapistudio.potterspells','co.april2019.pindia','org.rntcp.nikshay','com.puzzlegames.block.puzzle','com.superevilmegacorp.game','com.tinder','com.twitter.android','com.delhi.university.android','com.zaidfdgh.vikingsquiz','com.pl.cwc_2015'];
```

```python
len(appIds)
```

```python
def put_data(appId):
  return app(
    appId,
    lang='en',
    country='us'
    );
```

```python
apps = [];
for appId in appIds:
  apps.append(put_data(appId));
```

```python
redundant_columns=['description','descriptionHTML','screenshots','histogram','comments'];
```

```python
def remove_columns(app):
  for column in redundant_columns:
    del app[column];
```

```python
for app in apps:
  remove_columns(app);
```

```python
df=pd.json_normalize(apps)
```
